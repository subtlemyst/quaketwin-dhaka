# Phase 1 — Hazard Pipeline

Generate shaking, amplification, and liquefaction maps for Mw 7.2 Dauki scenario.

```python
import json
import sys
from pathlib import Path

ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
sys.path.insert(0, str(ROOT / "src"))

from quaketwin.pipeline import run_phase1_hazard
from quaketwin.hazard.scenario import load_default_scenario

scenario = load_default_scenario()
fc = run_phase1_hazard(scenario=scenario)
out = ROOT / "data" / "processed" / "hazard_mw72_dauki.geojson"
out.parent.mkdir(parents=True, exist_ok=True)
with open(out, "w") as f:
    json.dump(fc, f, indent=2)
print(f"Wrote {len(fc['features'])} cells")
```

```python
import matplotlib.pyplot as plt

lons = [f["geometry"]["coordinates"][0] for f in fc["features"]]
lats = [f["geometry"]["coordinates"][1] for f in fc["features"]]
pga = [f["properties"]["amplified_pga_g"] for f in fc["features"]]
liq = [f["properties"]["liquefaction_index"] for f in fc["features"]]

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].scatter(lons, lats, c=pga, cmap="YlOrRd", s=8)
axes[0].set_title(f"Amplified PGA — {scenario.id}")
axes[1].scatter(lons, lats, c=liq, cmap="Blues", s=8, vmin=0, vmax=1)
axes[1].set_title("Liquefaction index")
plt.tight_layout()
plt.savefig(ROOT / "data" / "processed" / "hazard_maps_preview.png", dpi=150)
plt.show()
```
